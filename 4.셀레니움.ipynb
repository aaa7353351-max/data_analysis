### ※ xpath

# **3. 픽사베이 이미지 다운로드** 



```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request,urlopen
```

```python
driver = webdriver.Chrome()
url = 'https://pixabay.com/ko/images/search/snow/'
driver.maximize_window() #창이 무조건 크게 실행됨
driver.get(url)
time.sleep(2)
```

```python
image_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[1]/div[7]/div/a/img'
wait = WebDriverWait(driver,10) # 못찾으면 최대 10초까지 기다려줌 
image_element = wait.until(EC.presence_of_element_located((By.XPATH,image_xpath)))
```

```python
image_url = image_element.get_attribute("src")
image_byte = Request(image_url, headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'})
image_byte
```

```python
with open('snow.jpg','wb') as f:
    f.write(urlopen(image_byte).read())
print("파일저장 완")

driver.quit()
```

# **야놀자 리뷰**
* [사이트](https://nol.yanolja.com/reviews/domestic/1000111510)

```python
def crawl_yanolja_reviews(name, url, max_scroll=3):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(5)
    last_height = driver.execute_script("return document.body.scrollHeight")
    # print(last_height)
    same_count = 0
    scroll_times = 0
    while True:
        scroll_times += 1
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            same_count += 1
        else:
            same_count = 0
            last_height = new_height
        if same_count >= 3:
            print('더 이상 새로운 리뷰가 없습니다.')
            break
        if scroll_times >= max_scroll:
            print('최대 스크롤 횟수 도달. 종료합니다.')
            break
    containers = driver.find_elements(By.XPATH,'/html/body/div[1]/section/div/div[5]/div')
    print('총 리뷰 개수:',len(containers))

    for c in containers:
        try:
            # raw_text = c.text.strip()
            date_text = c.find_element(By.XPATH,'div[2]/div/div[1]/div[2]/p').text
            comment_text = c.find_element(By.XPATH,'div[2]/div/div[3]/div[1]/p').text
            print(date_text)
            # div[2]/div/div[3]/div[1]/p
        except: 
            date_text=""
            comment_text = ""

        review_list.append({
            "review": comment_text,
            "date": date_text})
            
    driver.quit()
    return review_list
    # /html/body/div[1]/section/div/div[5]/div[60]
    
```

```python
crawl_yanolja_reviews("베니온","https://nol.yanolja.com/reviews/domestic/1000111510")
```
